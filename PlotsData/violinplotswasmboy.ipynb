```python
from matplotlib import pyplot as plt
import pandas as pd
import seaborn as sns
sns.set_style('whitegrid')
custom_palette=('#5B9BD5', '#A4CE88',)
sns.set_palette(custom_palette) 
```

```python
dfhp20 = pd.read_csv("../WasmBoy/BackToColor/backtocolorall.csv",error_bad_lines=False, encoding="ISO-8859-1")
dfhp20 = dfhp20.rename(columns={'Package': 'Energy (Joules)', 'Time': 'Time (Seconds)'})
dfhp40 = pd.read_csv("../WasmBoy/DinosOfflineAdventure/dinosofflineadventureall.csv",error_bad_lines=False, encoding="ISO-8859-1")
dfhp40 = dfhp40.rename(columns={'Package': 'Energy (Joules)', 'Time': 'Time (Seconds)'})
dfhp80 = pd.read_csv("../WasmBoy/Pokemon/pokemonall.csv",error_bad_lines=False, encoding="ISO-8859-1")
dfhp80 = dfhp80.rename(columns={'Package': 'Energy (Joules)', 'Time': 'Time (Seconds)'})
dfpaper = pd.read_csv("../WasmBoy/SuperMario/supermarioall.csv",error_bad_lines=False, encoding="ISO-8859-1")
dfpaper = dfpaper.rename(columns={'Package': 'Energy (Joules)', 'Time': 'Time (Seconds)'})
dfppt = pd.read_csv("../WasmBoy/SuperMarioLand/supermariolandall.csv",error_bad_lines=False, encoding="ISO-8859-1")
dfppt = dfppt.rename(columns={'Package': 'Energy (Joules)', 'Time': 'Time (Seconds)'})
dfppt2 = pd.read_csv("../WasmBoy/TobuTobuGirl/tobutobugirlall.csv",error_bad_lines=False, encoding="ISO-8859-1")
dfppt2 = dfppt2.rename(columns={'Package': 'Energy (Joules)', 'Time': 'Time (Seconds)'})
```

```python
fig_dims = (12, 6)
fig, ax = plt.subplots(1,2,num=1,figsize=fig_dims)
sns.violinplot(x='Browser', y='Energy (Joules)',hue='Language', data=dfhp20, linewidth=1,ax=ax[0]).set_title("BackToColor - Energy",weight='bold').set_fontsize('16');
sns.violinplot(x='Browser', y='Time (Seconds)',hue='Language', data=dfhp20, linewidth=1,ax=ax[1]).set_title("BackToColor - Time",weight='bold').set_fontsize('16');
```
